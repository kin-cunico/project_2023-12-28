### renaming columns for better understanding

```python
columns_dict = {
    'timestamp': 'date',
    'cnt': 'total',
    't1': 'temperature',
    't2': 'feels_like_temperature',
    'hum': 'humidity',
    'wind_speed': 'wind_speed',
    'weather_code': 'weather',
    'is_holiday': 'is_holiday',
    'is_weekend': 'is_weekend',
    'season': 'season'
}

bikes.rename(columns=columns_dict, inplace=True)

bikes.head()
```

## transforming data

```python
bikes.humidity = bikes.humidity / 100

bikes.head()
```

```python
bikes.season = bikes.season.astype('string')
bikes.weather = bikes.weather.astype('string')


season_dict = {
    '0.0': 'spring',
    '1.0': 'summer',
    '2.0': 'fall',
    '3.0': 'winter'
}

bikes.season = bikes.season.map(season_dict)

weather_dict = {
    '1.0': 'clear',
    '2.0': 'mist',
    '3.0': 'light_rain',
    '4.0': 'heavy_rain',
    '7.0': 'thunderstorm',
    '10.0': 'freezing',
    '26.0': 'snowfall'
    
}

bikes.weather = bikes.weather.map(weather_dict)

bikes.info()

bikes.head()
```
